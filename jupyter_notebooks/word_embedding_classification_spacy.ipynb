## Text einlesen

```python
import csv
from string import punctuation
with open("./test.csv", "r", encoding="utf-8") as f:
    content = f.readlines()

sentences = []
for line in content[:100]:
    sentence = line.split(";")[1]
    sentences.append(sentence)

norm_corpus = [[word for word in sent if word not in punctuation] for sent in sentences]
norm_corpus = [''.join(tok_sent) for tok_sent in norm_corpus]
norm_corpus
```

## Text klassifizieren

```python
import spacy
nlp = spacy.load('de_core_news_md')
```

```python
nlp_sentences = [nlp(sentence) for sentence in norm_corpus]
sentences = [sentence for sentence in norm_corpus]
```

```python
import pandas
similar = []
for sentence1 in nlp_sentences:
    similar_sentence = []
    for sentence2 in nlp_sentences:
        similar_sentence.append(sentence1.similarity(sentence2))
    similar.append(similar_sentence)

similar_df = pandas.DataFrame(similar)
similar_df
```

```python
most_similar_sentences = similar_df[similar_df < 1].max()
most_similar_sentences
```
